```python
import pandas as pd 
import numpy as np

#读取数据，了解基本情况
df = pd.read_excel('data.xlsx',index_col=0)
print("数据表的行列为：",df.shape)
print('获取列名\n',df.columns)

#各列数据类型和，空值情况
df.info()

#数值型数据的描述性分布
df.describe()
df.head()
```

```python
#开始处理G列——即“期望薪资”

#G列填充空值为‘无’
#注意到只有1407个数据，说明可能存在其他形式的null，比如na，nan
e_salary = df['期望薪资']
print(e_salary.isnull().sum()) # 有23个空值，1407加23正好1430
e_salary = e_salary.fillna('无')
print(e_salary.isnull().sum())


import re
#找到薪资单位不同的行索引
index_year= e_salary[e_salary.str.contains('年')].index
index_10000= e_salary[e_salary.str.contains('万')].index
print(index_year)
print(index_10000)

#切分出薪资数字
num_salary = e_salary.str.findall('^\d+-\d+')
num_salary=num_salary.astype("str")
print(type(num_salary[1]))#元素为str

#划分薪资上限下限
low = pd.Series()
high = pd.Series()
low,high = num_salary.str.split('-',1).str
#[''],需要反斜杠转义
low = low.str.split("\[\'").str[1]
high = high.str.split("\'\]").str[0]
# print(low,high)
df['low'] = low.astype('float')
df['high'] = high.astype('float')

# 汉字单位统一为元/月,根据年&万单位的索引进行统一
df.loc[index_10000,['low','high']]= df.loc[index_10000,['low','high']]*10000
df.loc[index_year,['low','high']] = df.loc[index_year,['low','high']]/12

# 得到期望薪资上下限，单位元/月，
# print(df['low'],df['high'])
print(np.isnan(df['low']))
```
